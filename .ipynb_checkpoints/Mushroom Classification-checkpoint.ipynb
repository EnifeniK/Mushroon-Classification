```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('mushroom_classification/mushrooms.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
sns.countplot(x='class', data=df)
```

```python
df.columns
```

```python
the_dummies = pd.get_dummies(df[['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']], drop_first=True)
```

```python
new_df = df.drop(['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat'], axis=1)
new_df
```

```python
df = pd.concat([new_df, the_dummies], axis=1)
```

```python
df
```

```python
df['class'] = df['class'].replace('p', 1).replace('e', 0)
```

```python
df
```

```python
#Training the model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
the_model = LogisticRegression()
```

```python
the_model.fit(X_train, y_train)
```

```python
predictions = the_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
second_model = DecisionTreeClassifier()
```

```python
second_model.fit(X_train, y_train)
```

```python
pred = second_model.predict(X_test)
```

```python
print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```

```python
test_score = second_model.score(X_test, y_test)
test_score
```
